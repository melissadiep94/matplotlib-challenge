## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
# Combine the data into a single dataset
combined_data = pd.merge(study_results, mouse_metadata, on=["Mouse ID"])
combined_data
# Display the data table for preview
```

```python
# Checking the number of mice.
num_mice = combined_data.nunique()["Mouse ID"]
num_mice

num_mice_df = pd.DataFrame({"Total # of Mice":[num_mice]})
num_mice_df
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicateDFRow = combined_data[combined_data.duplicated()]
print(duplicateDFRow)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

df_duplicate_mouseID = combined_data[combined_data['Mouse ID'] == "g989"]
df_duplicate_mouseID
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
#combined_data.drop([combined_data['Mouse ID'] == "g989"])

df_clean = combined_data[combined_data['Mouse ID'] != "g989"]
df_clean = pd.DataFrame(df_clean)
df_clean
```

```python
# Checking the number of mice in the clean DataFrame.
num_mice_clean = df_clean.nunique()["Mouse ID"]
num_mice_clean 

num_mice_df_clean = pd.DataFrame({"Total # of Mice":[num_mice_clean ]})
num_mice_df_clean
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 


df_simp = df_clean[['Drug Regimen','Tumor Volume (mm3)']]

df_mean  = df_simp.groupby("Drug Regimen").mean()
df_mean = pd.DataFrame(df_mean)
df_mean = df_mean.rename(columns = {'Tumor Volume (mm3)' : 'Mean Tumor Volume (mm3)'}) 

df_median = df_simp.groupby("Drug Regimen").median()
df_median = pd.DataFrame(df_median)
df_median = df_median.rename(columns = {'Tumor Volume (mm3)' : 'Median Tumor Volume (mm3)'}) 

df_var = df_simp.groupby("Drug Regimen").var()
df_var = pd.DataFrame(df_var)
df_var = df_var.rename(columns = {'Tumor Volume (mm3)' : 'Variance Tumor Volume (mm3)'}) 


df_std = df_simp.groupby("Drug Regimen").std()
df_std = pd.DataFrame(df_std)
df_std = df_std.rename(columns = {'Tumor Volume (mm3)' : 'Std. Tumor Volume (mm3)'}) 

df_sem = df_simp.groupby("Drug Regimen").sem()
df_sem = pd.DataFrame(df_sem)
df_sem = df_sem.rename(columns = {'Tumor Volume (mm3)' : 'SEM Tumor Volume (mm3)'}) 


# Assemble the resulting series into a single summary dataframe.

df_mean_median = pd.merge(df_mean, df_median, on='Drug Regimen')
df_mean_median_var = pd.merge(df_mean_median, df_var, on='Drug Regimen') 
df_mean_median_var_std = pd.merge(df_mean_median_var, df_std, on='Drug Regimen') 
df_consolidated = pd.merge(df_mean_median_var_std, df_sem, on='Drug Regimen') 
df_consolidated 
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

stats_agg = df_simp.groupby("Drug Regimen").agg(["mean", "median", "var", "std", "sem"])
stats_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

drugs_data_df = df_clean.set_index("Drug Regimen")
drug_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
selected_drugs_df=drugs_data_df.loc[drug_list] 

# Start by getting the last (greatest) timepoint for each mouse
mouse_group_df=selected_drugs_df.groupby(["Mouse ID"])

timepoint_max_df =pd.DataFrame(mouse_group_df["Timepoint"].max())
timepoint_max_df.head()

# Start by getting the last (greatest) timepoint for each mouse



```

```python
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
re_selected_drugs_df=selected_drugs_df.reset_index()

merged_timepoint_df= pd.merge(timepoint_max_df, re_selected_drugs_df[["Tumor Volume (mm3)","Mouse ID","Timepoint", "Drug Regimen"]], on=["Mouse ID","Timepoint"])
merged_timepoint_df.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)

tumor_volume = []

indexed_drugs_df = merged_timepoint_df.set_index(["Drug Regimen"])
# Create empty list to fill with tumor vol data (for plotting)



    
    
    
for drug in drug_list:
    print(f"\n----- {drug} ------")
    drug_df = indexed_drugs_df.loc[[drug]]
    # looping through tumor volume data to fill for plotting  
    tumor_volume.append(drug_df["Tumor Volume (mm3)"])
    
    #Calculate the IQR
    quartiles = drug_df["Tumor Volume (mm3)"].quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq

    print(f"The lower quartile of tumor volume is: {lowerq}")
    print(f"The upper quartile of tumor volume is: {upperq}")
   
    print(f"The interquartile range of tumor volume is: {iqr}")
    
    # determine if there are any potential outliers. 
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f"Values below {lower_bound} or above {upper_bound} could be outliers.")
    # printing potential outliers
    outlier_tumor_volume = drug_df.loc[(drug_df["Tumor Volume (mm3)"] < lower_bound) | (drug_df["Tumor Volume (mm3)"] > upper_bound)]
    if outlier_tumor_volume.empty == False:
        print("Potential ouliers:")
        print(outlier_tumor_volume)
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig1, ax1 = plt.subplots()
ax1.set_ylabel("Final Tumor Volume (mm3)")
ax1.boxplot(tumor_volume,flierprops=dict(markerfacecolor='g', marker='8'))
plt.xticks([1,2,3,4], drug_list)
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
Capomulin_df=drugs_data_df.loc["Capomulin"]
re_Capomulin_df=Capomulin_df.reset_index()
re_Capomulin_df 

re_Capomulin_df_mouseID  = re_Capomulin_df.groupby("Mouse ID").mean()
re_Capomulin_df_mouseID = re_Capomulin_df_mouseID[['Tumor Volume (mm3)', 'Weight (g)']].copy()
re_Capomulin_df_mouseID
```

```python
plt.scatter(re_Capomulin_df_mouseID.iloc[:,1],re_Capomulin_df_mouseID.iloc[:,0])
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
tumor_vol = re_Capomulin_df_mouseID.iloc[:,0]
weight = re_Capomulin_df_mouseID.iloc[:,1]
(slope, intercept, rvalue, pvalue, stderr) = linregress(weight, tumor_vol)
regress_values = weight * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(weight,regress_values,"r-")
correlation = st.pearsonr(weight, tumor_vol)
print(f"The r-squared is: {rvalue**2}")
print(f"The correlation between Tumor Vol and Weight for Capomulin tests is {round(correlation[0],2)}")
plt.show()
```
